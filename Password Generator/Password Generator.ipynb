```python
import tkinter as tk
from tkinter import messagebox
import random
import string
```

```python
def generate_password():
    try:
        length = int(length_entry.get())
        if length <= 0:
            messagebox.showwarning("Warning", "Password length must be greater than zero.")
            return

        lowercase = "abcdefghijklmnopqrstuvwxyz"
        uppercase = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
        digits = "0123456789"
        punctuation = "!@#$%^&*()_+-=[]{}|;:,.<>?"

       
        all_chars = ""
        if lowercase_var.get():
            all_chars += lowercase
        if uppercase_var.get():
            all_chars += uppercase
        if digits_var.get():
            all_chars += digits
        if punctuation_var.get():
            all_chars += punctuation

        if not all_chars:
            messagebox.showwarning("Warning", "Please select at least one character type.")
            return

       
        password = ''.join(random.sample(all_chars, length))
        password_entry.delete(0, tk.END)
        password_entry.insert(tk.END, password)

    except ValueError:
        messagebox.showwarning("Warning", "Please enter a valid number for password length.")
```

```python
root = tk.Tk()
root.title("Enhanced Password Generator")
root.geometry("400x400")
root.configure(bg='#F0F4C3')  # Light yellow background

title_label = tk.Label(root, text="Password Generator", font=("Arial", 24), bg='#F0F4C3', fg='#388E3C') 
title_label.pack(pady=10)

length_label = tk.Label(root, text="Password Length:", font=("Arial", 14), bg='#F0F4C3')
length_label.pack()

length_entry = tk.Entry(root, width=10, font=("Arial", 14))
length_entry.pack(pady=5)

lowercase_var = tk.IntVar()
lowercase_check = tk.Checkbutton(root, text="Lowercase", variable=lowercase_var, font=("Arial", 12), bg='#F0F4C3')
lowercase_check.pack(anchor=tk.W)

uppercase_var = tk.IntVar()
uppercase_check = tk.Checkbutton(root, text="Uppercase", variable=uppercase_var, font=("Arial", 12), bg='#F0F4C3')
uppercase_check.pack(anchor=tk.W)

digits_var = tk.IntVar()
digits_check = tk.Checkbutton(root, text="Digits", variable=digits_var, font=("Arial", 12), bg='#F0F4C3')
digits_check.pack(anchor=tk.W)

punctuation_var = tk.IntVar()
punctuation_check = tk.Checkbutton(root, text="Punctuation", variable=punctuation_var, font=("Arial", 12), bg='#F0F4C3')
punctuation_check.pack(anchor=tk.W)

generate_button = tk.Button(root, text="Generate Password", font=("Arial", 14), bg='#FFA726', fg='white', command=generate_password)  
generate_button.pack(pady=10)

password_label = tk.Label(root, text="Generated Password:", font=("Arial", 14), bg='#F0F4C3')
password_label.pack()

password_entry = tk.Entry(root, width=30, font=("Arial", 14))
password_entry.pack(pady=5)

root.mainloop()
```
